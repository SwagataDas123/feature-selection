```python
def func(sona):

  #data preprocessing
  label_encoder = preprocessing.LabelEncoder()
  #sona.rename(columns={'class ': 'class'}, inplace=True)#for pageblocks dataset
  sona['class']=label_encoder.fit_transform(sona['class'])
  """
  #for weather2 dataset
sona['outlook']=label_encoder.fit_transform(sona['outlook'])
sona['windy']=label_encoder.fit_transform(sona['windy'])

  """
  #sona['Marital Status']= label_encoder.fit_transform(sona['Marital Status'])#for customers1 dataset
  #sona.head(5)
  sona.info()#checking null values
  #sona=sona.dropna()#removing null values(for vehicle1 dataset)
  """
  #for dermatology datset
  column_name = 'Age'

  #Specify the string you want to check for search_string = "?"
  if '?' in sona[column_name].astype(str).values:
  print(f"The column '{column_name}' contains '?'")
  sona['Age'] = pd.to_numeric(sona['Age'], errors='coerce')

  #Identify rows where the numeric column contains '?'
  rows_with_question_mark = sona['Age'].isna()

  #Drop rows with '?' in the numeric column
  sona = sona[~rows_with_question_mark]

sona['Age'] = pd.to_numeric(sona['Age'], errors='coerce')
  """
  X=sona.drop("class",axis=1)
  y=sona["class"]
  (sona < 0).any().any()#negative value in dataset

  #picking up 20% of the columns
  k2=(20*(sona.shape[1]))/100.0
  k1=math.ceil(k2)

  #CORRELATION
  X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)#train test split
  X_train.shape,X_test.shape

  """#NORMALISATION
  sc=StandardScaler()
  X_train=sc.fit_transform(X_train)
  X_test=sc.transform(X_test)"""

  #correlation function
  def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
      for j in range(i):
        if(corr_matrix.iloc[i,j]>threshold):
          colname=corr_matrix.columns[i]
          col_corr.add(colname)
    return col_corr
    #correlation function ends

  corr_features=correlation(X_train,0.8)#passing  parameters to correlation function
  corr_features

  new_dataset=sona.drop(corr_features,axis=1)#drop those features having correlation greater than 0.8 threshold
  #new dataset
  new_dataset.head(5)

  #sorting the correlations list
  corr_matrix=new_dataset.corr()
  ch1=corr_matrix["class"].tolist()
  ch=corr_matrix.columns.tolist()
  for i in range(len(ch)-1):
    for j in range(len(ch)-i-1):
      if(ch1[j+1]>ch1[j]):
        ch1[j],ch1[j+1]=ch1[j+1],ch1[j]
        ch[j],ch[j+1]=ch[j+1],ch[j]
  cfd=ch[1:k1+1]#picking up 20% of the columns
  #print(cfd)

  #MUTUAL INFORMATION
  mutual_info=mutual_info_classif(X_train,y_train)
  mutual_info
  mutual_info=pd.Series(mutual_info)
  mutual_info.index=X_train.columns
  mutual_info.sort_values(ascending=False)#sorting

  #selecting features
  sel_k_columns=SelectKBest(mutual_info_classif,k=k1)
  sel_k_columns.fit(X_train,y_train)
  mifd=list(X_train.columns[sel_k_columns.get_support()])
  #print(mifd)

  cfd=set(cfd)
  mifd=set(mifd)
  #common columns name
  z=cfd.intersection(mifd)
  #print(z) #print(len(z))

  #similarity between MI and CORRELATION
  sp=len(z)/(len(cfd)+len(mifd)-len(z))
  sp=sp*100
  #print(sp*100)

  #Chi Square Test
  f_p_values=chi2(abs(X_train),y_train)
  f_values=pd.Series(f_p_values[0])#f values
  f_values.index=X_train.columns
  p_values=pd.Series(f_p_values[1])#p values
  p_values.index=X_train.columns

  #picking up 20% of the columns and selecting features
  sel_kk_columns=SelectKBest(score_func=chi2,k=k1)
  fit=sel_kk_columns.fit(abs(X_train),y_train)

  c2fd=list(X_train.columns[sel_kk_columns.get_support()])

  c2fd=set(c2fd)
  #mifd=set(mifd)
  #common columns name
  z1=c2fd.intersection(mifd)
  #print(z)
  #print(len(z))

  #similarity between MI and CHI2
  sp1=len(z1)/(len(c2fd)+len(mifd)-len(z1))
  #print(sp*100)
  sp1=sp1*100

  #common columns name
  z2=cfd.intersection(c2fd)

  #similarity between CORRELATION and CHI2
  sp2=len(z2)/(len(c2fd)+len(cfd)-len(z2))
  sp2=sp2*100

  #print(sp,sp1)
  return sp,sp1,sp2#returning similarity
  #function ends
```
